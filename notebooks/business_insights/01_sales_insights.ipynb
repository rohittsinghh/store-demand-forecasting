```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../../data/raw/train.csv", parse_dates=["date"])
df = df.sort_values("date")

df.head()
```

```python
store_perf = df.groupby("store")["sales"].mean().sort_values(ascending=False)

store_perf.plot(
    kind="bar",
    figsize=(10,4),
    title="Average Sales per Store"
)
plt.ylabel("Average Sales")
plt.show()
```

```python
df["day_type"] = df.date.dt.dayofweek.apply(
    lambda x: "Weekend" if x >= 5 else "Weekday"
)

df.groupby("day_type")["sales"].mean().plot(
    kind="bar",
    title="Weekend vs Weekday Sales"
)
plt.ylabel("Average Sales")
plt.show()
```
